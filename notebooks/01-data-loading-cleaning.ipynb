Now, let's fill in missing data and save those to csv files by region.

```python
def clean_and_process_covid_data(df):
    df["date"] = pd.to_datetime(df["date"])
    df.drop(columns=["geocoded_state"], inplace=True) # drop this column because it's empty for all rows
    df.set_index("date", inplace=True)
    df.drop(columns=["Unnamed: 0"], inplace=True) # drop the original index
    df = df.sort_index()

    # Fill in missing dates
    full_range = pd.date_range(start=df.index.min(), end=df.index.max(), freq="D")
    missing = set(full_range) - set(df.index)
    if len(missing) > 0:
        df = df.asfreq("D")

    # Fill in missing state values
    if df["state"].isna().sum() > 0:
        df["state"] = state

    # Fill in the missing data values using a forward fill
    # then do a backward fill
    df_clean = df.copy()
    df_clean = df_clean.ffill()
    df_clean = df_clean.bfill()

    return df_clean
```

```python
def add_urban_ratio_and_region(df, ratios, region):
    df["region"] = region
    df["urban_to_total_ratio"] = 0
    for state in ratios.keys():
        sel = df["state"] == state
        df.loc[sel, "urban_to_total_ratio"] = ratios[state]
    return df
```

```python
def make_region_dataframe(states, urban_ratios, region):
    df = pd.DataFrame()
    for state in states:
        state_df = pd.read_csv(f"../data/covid_adult_data_raw_{state}.csv")
        state_df_clean = clean_and_process_covid_data(state_df)
        state_df_full = add_urban_ratio_and_region(state_df_clean, urban_ratios, region)
        df = pd.concat([df, state_df_full], axis=0)
    return df
```

State divisions are were determined by [this source](https://www.fla-shop.com/resources/us-regions/). Ratio of urban population to total popuation from [Wikipedia](https://en.wikipedia.org/wiki/Urbanization_in_the_United_States#cite_note-6).
